# Interpolação

Frequentemente você terá de fazer estimativas de valores intermediários entre dados discretos. O método mais comum usado para esse propósito é a interpolação polinomial. A fórmula geral para um polinômio de grau (n-1) pode ser escrita:

$$f(x) = a_0 + a_1x+a_2x^2+a_3x^3 + ... + a_nx^n$$

Para $n$ pontos dados, existe um, e somente um, polinômio (n-1) que passa por todos os pontos. Por exemplo, existe uma única reta (isto é, um polinômio de primeiro grau) que liga dois pontos (Figura a). Analogamente, existe uma única parábola ligando um conjunto de três pontos (Figura b). A _interpolação polinomial_ consiste em determinar o único polinômio de grau (n-1) que passa pelos pontos $n$ dados. Esse polinômio, então, fornece uma fórmula para calcular valores intermediários.

## Determinando os coeficientes do polinômio

### O método direto

Uma maneira simples de calcular os coeficientes de um polinômio é baseada no fato de que $n$ pontos dados são necessários para determinar $n$ coeficientes. Como no exemplo a seguir, isso permite gerar $n$ equações algébricas lineares que podem ser resolvidas simultaneamente para calcular os coeficientes.

#### Exemplo 1:

Dado os pares (x,y), determine P(4):

x | y
---|---
3|40
5|82
9|214

##### Interpolação Linear

Para interpolação linear precisamos obter um polinômio com a seguinte forma:

$$P_1(x) = a_0 + a_1x$$

Temos dois coeficientes para determinar ($a_0$ e $a_1$) para isso precisamos de uma sistema com duas equações. De acordo com a tabela fornecida e com o que foi pedido, P(4), escolhemos os pontos mais próximos de $x=4$ e obtemos:

$$P_1(3) = a_0 + 3a_1 = 40$$
$$P_1(5) = a_0 + 5a_1 = 82$$

montamos o sistema:

$$\begin{bmatrix}
1 & 3\\
1 & 5
\end{bmatrix}
\begin{bmatrix}
a_0\\
a_1\\
\end{bmatrix}
=
\begin{bmatrix}
40\\
82
\end{bmatrix}$$

Resolvendo o sistema através do método de Gauss obtemos um novo sistema:

$$\begin{bmatrix}
1 & 3\\
0 & 2
\end{bmatrix}
\begin{bmatrix}
a_0\\
a_1\\
\end{bmatrix}
=
\begin{bmatrix}
40\\
42
\end{bmatrix}$$

Podemos então determinar os coeficientes:

$$a_1 = 21$$

$$a_0 = -23$$

O polinômio e o seu domínio:

$$P_1(x) = -23 + 21x$$

$$ D = \{ x \in \mathscr{R} | 3 \leq x \leq 5 \} $$

e por fim:

$$P_1(4) = 61 $$

##### Interpolação Quadrática

Agora para interpolação quadrática precisamos obter um polinômio com a seguinte forma:

$$P_2(x) = a_0 + a_1x + a_2x^2$$

por isso utilizamos 3 pontos para criar o sistema.

$$P_2(3) = a_0 + 3a_1 + 9a_2 = 40$$
$$P_2(5) = a_0 + 5a_1 + 25a_2 = 82$$
$$P_2(9) = a_0 + 9a_1 + 81a_2 = 214$$

e montamos o sistema:

$$\begin{bmatrix}
1 & 3 & 9\\
1 & 5 & 25\\
1 & 9 & 81
\end{bmatrix}
\begin{bmatrix}
a_0\\
a_1\\
a_2
\end{bmatrix}
=
\begin{bmatrix}
40\\
82\\
214
\end{bmatrix}$$

Resolvendo o sistema através do método de Gauss com pivotação parcial obtemos um novo sistema:

$$\begin{bmatrix}
1 & 3 & 9\\
0 & 6 & 72\\
0 & -8 & -16
\end{bmatrix}
\begin{bmatrix}
a_0\\
a_1\\
a_2
\end{bmatrix}
=
\begin{bmatrix}
40\\
174\\
-16
\end{bmatrix}$$

Podemos então determinar os coeficientes:

$$a_2 = 2$$

$$a_1 = 5$$

$$a_0 = 7$$

O polinômio e o seu domínio:

$$P_2(x) = 7 + 5x + 2x^2$$

$$ D = \{ x \in \mathscr{R} | 3 \leq x \leq 9 \} $$

e por fim:

$$P_2(4) = 59 $$

## Restrições

Embora a abordagem do método direto forneça uma maneira fácil de efetuar interpolação, ela tem a um deficiência séria. Para compreender essa falha, observe que a matriz de coeficientes tem sempre uma estrutura determinada:

$$\begin{bmatrix}
1 & x_1 & x_1^2\\
1 & x_2 & x_2^2\\
1 & x_3 & x_3^2
\end{bmatrix}
\begin{bmatrix}
a_0\\
a_1\\
a_2
\end{bmatrix}
=
\begin{bmatrix}
f(x_1)\\
f(x_2)\\
f(x_3)
\end{bmatrix}$$

Matrizes de coeficientes dessa forma são chamadas de _matrizes de [Vandermonde]_(https://pt.wikipedia.org/wiki/Matriz_de_Vandermonde). Essas matrizes são muito [mal condicionadas](https://en.wikipedia.org/wiki/Condition_number) ; ou seja, suas soluções são sensíveis a erros de arredondamento. Como consequência, há abordagens alternativas que não apresentam essa deficiência.
